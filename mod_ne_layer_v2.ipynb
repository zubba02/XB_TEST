```python
import fiona
import shapely.geometry
import numpy as np
import geopandas as gpd
```

```python
x_grid = np.loadtxt('x.grd')
y_grid = np.loadtxt('y.grd')
bed_grid = np.loadtxt('bed.dep')
```

```python
x = np.ravel(x_grid)
y = np.ravel(y_grid)
bed = np.ravel(bed_grid)
```

```python
# Read the shapefile containing multiple polygons
gdf = gpd.read_file('polyg.shp')

ne_layer = []

for i,j,k in zip (x,y,bed):

    point = shapely.geometry.Point(i, j)
    
    is_inside_any_polygon = any(point.within(polygon) for polygon in gdf['geometry'])
    
    if is_inside_any_polygon:
        print("The point is inside one polygon in the shapefile.")
        ne_layer.append(abs(k))
    else:
        ne_layer.append(0)
        print("The point is outside all polygons in the shapefile.")
        
```

```python
ne_layer = np.array(ne_layer)
```

```python
ne_layer = np.reshape(ne_layer, (x_grid.shape[0],x_grid.shape[1]))
```

```python
np.savetxt('mod_ne_layer.grd',  ne_layer)
```

```python
np.savetxt('All.csv', np.column_stack((x,y,ne_layer.ravel())))
```
